```python
import csv
import time
from datetime import datetime
from psychopy import visual, event, core
from screeninfo import get_monitors
import os
import os.path as osp
import re

# Get the computer screen size from hardware info and set the window size to that
screenMetadata = get_monitors()[0]
screenSize = (screenMetadata.width, screenMetadata.height)
print(screenSize)




# Variabile per il nome del soggetto (CAMBIARE)
name = "soggetto"
  
# Read texts from txt files in "texts" folder. CAMBIARE
textsFolder = osp.join(".", "texts2")

 



# Funzione per estrarre numeri dai nomi dei file, utile per ordinamento numerico
def get_text_number(filename):  
    match = re.search(r'\d+', filename)  # Cerca il primo numero nel nome del file
    return int(match.group()) if match else float('inf')

# Leggi i testi dalla cartella "texts" in ordine numerico
texts = []
for textFilename in sorted(os.listdir(textsFolder), key=get_text_number):
    if textFilename.endswith(".txt"):
        with open(osp.join(textsFolder, textFilename), "r", encoding="utf-8") as file:
            texts.append("\n\n\n\n"+ file.read())
```

```python
diametro_originale_cm = 7.5
larghezza_originale_px = 1920
altezza_originale_px = 1080
larghezza_nuova_px = 1366
altezza_nuova_px = 768

# Calcolo del fattore di scala medio
scala_media = 0.5 * ((larghezza_nuova_px / larghezza_originale_px) + (altezza_nuova_px / altezza_originale_px))

# Calcolo del nuovo diametro dello scotoma
diametro_nuovo_cm = diametro_originale_cm * scala_media
print(diametro_nuovo_cm)

```

```python
# TextBox Parameters
lineSpacing = 0.77
letterHeight = 50
textColor = [0, 0, 0]
backgroundColor = [-1, -1, -1]     
size = (1000, 500)

# Create a window with maximized size
win = visual.Window(size=screenSize, color=backgroundColor, units='pix', fullscr=True)

# Show an initial screen with the text "START"
start_screen = visual.TextStim(
    win=win,
    text='START',
    font='Arial',
    pos=(0, 0),
    height=letterHeight,  # Adjust as needed
    color=[1, 1, 1],  # White color for the text 
    units='pix'
)
start_screen.draw()
win.flip()

# Prepare to store timestamps
timestamps = []

# Wait for the first space key press to start
event.waitKeys(keyList=['space'])
timestamps.append(("Start", datetime.now().strftime("%Y-%m-%d %H:%M:%S.%f"))) 

# Nascondi il puntatore del mouse
win.mouseVisible = False










# Iterate through each text in the list
for i, text_to_show in enumerate(texts):

    # Create a black screen with a dash indicator
    dash = visual.TextStim(
        win=win,
        text='-',
        font='Arial',
        pos=(-500 + letterHeight / 2, 0),  # Adjust the position according to where the first letter will be
        height=letterHeight,  # Size of the dash
        color=[1, 1, 1],  # White color for the dash
        units='pix'
    )

    # Draw the black screen with the dash and display it
    win.color = backgroundColor  # Set the window background to black
    dash.draw()
    win.flip()

    # Wait for space key press to proceed
    event.waitKeys(keyList=['space'])

    # Create a TextBox2 for the current text
    text_box = visual.TextBox2(
        win=win,
        text=text_to_show,
        font='Arial',
        letterHeight=letterHeight,  # Font size
        color=textColor,  # Font color
        colorSpace='rgb',
        size=size,  # Adjust the size to fit your text
        pos=(0, 0),
        units='pix',
        alignment='center',  # Text alignment
        lineSpacing=lineSpacing  # Line spacing
    )

    # Draw and display the current TextBox2
    text_box.draw()
    win.flip()




     # Save the timestamp for when the text is displayed
    start_time = datetime.now()
    timestamps.append((f'Text {i + 1} Start', start_time.strftime("%Y-%m-%d %H:%M:%S.%f")))

    # Wait for space key press to proceed to the next text or escape to exit
    keys = event.waitKeys(keyList=['space', 'escape'])

    # Save the timestamp for when the text is finished
    end_time = datetime.now()
    timestamps.append((f'Text {i + 1} End', end_time.strftime("%Y-%m-%d %H:%M:%S.%f")))

    # Calculate the reading duration and save it
    duration = (end_time - start_time).total_seconds()
    timestamps.append((f'Text {i + 1} Duration', duration))

    # If 'escape' is pressed, exit the loop
    if 'escape' in keys:
        break








# Record the final timestamp when the task is completed
timestamps.append(("End", datetime.now().strftime("%Y-%m-%d %H:%M:%S.%f")))

# Record the initial timestamp when the task starts
start_time = datetime.now()   
timestamps.append(("Start", start_time.strftime("%Y-%m-%d %H:%M:%S.%f")))

# Determina la condizione in base alla cartella di testo selezionata
condition = "Texts1" if "texts1" in textsFolder else "Texts2"
# Definisce la struttura del percorso di salvataggio
save_path = f"C:\\Users\\robyt\\Documents\\ANALISI DATI\\Dati\\{name}\\Reading\\{condition}"
os.makedirs(save_path, exist_ok=True)  # Crea la directory se non esiste

# Record the initial timestamp when the task starts
start_time = datetime.now()
timestamps.append(("Start", start_time.strftime("%Y-%m-%d %H:%M:%S.%f")))

# Nome del file CSV con il nome del soggetto, la condizione e il timestamp di start
csv_filename = osp.join(save_path, f"{name}_Reading_{condition}_Pc.csv")


# Salva i timestamp nel CSV
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Event", "Timestamp"])
    for event_name, ts in timestamps:
        writer.writerow([event_name, ts])

# Stampa un messaggio con il nome del file
print(f"Timestamps saved to {csv_filename}")


# Close the window
win.close()

print(f"Timestamps saved to {csv_filename}")
```
